Also, based on RFM, K means clustering is performed and the optimal number of clusters is found using elbow curve and Davies Bouldin index. These clusters are further studied using mean values of recency, frequency, and the revenue generated.

##**Customer Segmentation**
Here we define three paramteres for each customer, namely- 

1. monetary value of the customer- total revenue
2. recency- time since last purchase
3. frequency- no of times shopped


This is commonly known as RFM analysis

Last invocie date in our data is  dec 9, 2011 which is to be used to calculate recency.

```python
pres=dt.date(2011,12,9)
df['Date']=(df['InvoiceDate']).dt.date
df.head()
```

```python
#forming recency table

recency=df.dropna().groupby('Customer ID',as_index=False)['Date'].max()
recency.columns=['Customer ID','Latest_purchase_date']

recency['Recency']=recency['Latest_purchase_date'].apply(lambda x:(pres-x).days)
recency.drop(columns=['Latest_purchase_date'],inplace=True)
recency.head()
```

```python
#forming monetary and frequency tables
df=df[df['Quantity']>0]
FM=df.groupby('Customer ID').agg({'Invoice':lambda x:len(x),
                                  'Revenue': lambda x:x.sum()})
FM.rename(columns={'Invoice':'Frequency',
                   'Revenue':'Money'},inplace=True)
FM.head()
```

```python
RFM=pd.merge(FM,recency,on='Customer ID')
RFM.head()
```

# **Now, we divide all three parameters into 5 equally spaced intervals. And rate each customer for all three paramteres. These rating are finally combined and an explicit function is defined (based on what these values represent) to give a label to each customer.**

```python
quant=RFM.quantile(q=[0.20,0.4,0.6,0.8])
quant=quant.to_dict()
quant
```

```python
segments=RFM.copy()
def Rec(x,feat,dict):
  if x<=dict[feat][0.2]:
    return 1
  elif x<=dict[feat][0.4]:
    return 2
  elif x<=dict[feat][0.6]:
    return 3
  elif x<=dict[feat][0.8]:
    return 4
  else:
    return 5

def FMscore(x,feat,dict):
  if x<=dict[feat][0.2]:
    return 5
  elif x<=dict[feat][0.4]:
    return 4
  elif x<=dict[feat][0.6]:
    return 3
  elif x<=dict[feat][0.8]:
    return 2
  else:
    return 1

segments['Rec_score']=segments['Recency'].apply(Rec,args=('Recency',quant))
segments['Freq_score']=segments['Frequency'].apply(FMscore,args=('Frequency',quant))
segments['Mon_score']=segments['Money'].apply(FMscore,args=('Money',quant))

segments.head()
```

```python
segments['RFM']=segments['Rec_score'].map(str)+segments['Freq_score'].map(str)+segments['Mon_score'].map(str)
segments.head()
```

```python
print('#Best customers ',len(segments.query(' (Rec_score==1) and (Freq_score==1) and (Mon_score==1 or Mon_score==2) ')))
print('#loyal less profitable ',len(segments.query(' (Rec_score==1) and (Freq_score==1) and (Mon_score==3 or Mon_score==4 or Mon_score==5) ')))
print('#Rare lost customers ',len(segments.query(' (Rec_score==4 or Rec_score==5) and (Freq_score==4 or Freq_score==5) and (Mon_score==4 or Mon_score==5) ')))
print('#lost profitable loyal customers ', len(segments.query( ' (Rec_score==4 or Rec_score==5) and (Freq_score==1 or Freq_score==2) and (Mon_score==1 or Mon_score==2) ')))
print('#new customers ',len(segments.query('  (Rec_score==1 or Rec_score==2) and (Freq_score==4 or Freq_score==5) ')))
```

```python
tag=[0]*len(segments)
for i in range(0,len(segments)):

  if (segments['RFM'][i]=='111') or (segments['RFM'][i]=='112') :
    tag[i]='Best customers'

  elif (segments['RFM'][i]=='113') or (segments['RFM'][i]=='114') or (segments['RFM'][i]=='115') :
    tag[i]='loyal less profitable'

  elif ((segments['Rec_score'][i]==1) or (segments['Rec_score'][i]==2)) and ((segments['Freq_score'][i]==4) or (segments['Freq_score'][i]==5)):
    tag[i]='new customers'

  elif ((segments['Rec_score'][i]==4) or (segments['Rec_score'][i]==5)) and ((segments['Freq_score'][i]==2) or (segments['Freq_score'][i]==1)) and ((segments['Mon_score'][i]==2) or (segments['Mon_score'][i]==1)):
    tag[i]='lost profitable loyal customers'

  elif ((segments['Rec_score'][i]==4) or (segments['Rec_score'][i]==5)) and ((segments['Freq_score'][i]==4) or (segments['Freq_score'][i]==5)) and ((segments['Mon_score'][i]==2) or (segments['Mon_score'][i]==1)) :
      tag[i]='Rare lost customers'

  else:
    tag[i]='others'
  
```

**The graph below shows how many customers fall in each cateogry. This is useful to get an idea about how the business is performing as a whole.**

```python
segments['tag']=tag
fig=segments.groupby('tag')['Customer ID'].nunique().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,15))
fig.drop([0],inplace=True)
sns.barplot(data=fig,x='tag',y='Customer ID')
```

## **# K-Means** **Clustering**

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.metrics import davies_bouldin_score

segments1=segments[['Rec_score',	'Freq_score',	'Mon_score']]
scaler=StandardScaler()
scaler.fit(segments1)
scaled=scaler.transform(segments1)
scaled=pd.DataFrame(scaled, columns=segments1.columns)
scaled.head()
```

```python
x=np.asarray(scaled)
sos_dist=[]
K=range(1,10)
for k in K:
  km=KMeans(n_clusters=k)
  km=km.fit(x)
  sos_dist.append(km.inertia_)
plt.plot(K,sos_dist,'bx-')
plt.xlabel('no fo clusters')
plt.ylabel('Sum of Sq. Distances')
plt.title('Elbow method to choose optimal no of clusters')
plt.show()
```

```python
def kmeans(normalised,clusters,orig):
  kmeans=KMeans(n_clusters=clusters,random_state=1)
  kmeans.fit(normalised)
  cluster_label=kmeans.labels_
  new_df=orig.assign(Cluster=cluster_label)
  model=TSNE(random_state=1)
  transformed=model.fit_transform(new_df)
  plt.title('view of {} clusters'.format(clusters))
  sns.scatterplot(x=transformed[:,0],y=transformed[:,1],hue=cluster_label,palette='RdYlGn_r',style=cluster_label)
  return new_df
plt.figure(figsize=(10,15))
plt.subplot(3,1,1)
k3=kmeans(scaled,3,RFM)
plt.subplot(3,1,2)
k4=kmeans(scaled,4,RFM)
plt.subplot(3,1,3)
k5=kmeans(scaled,5,RFM)
plt.tight_layout
```

**Bouldin metric  to find the optimal number of clusters**

```python
#score
kmeans=KMeans(n_clusters=3)
kmeans.fit(x)
print(davies_bouldin_score(x,kmeans.labels_))
kmeans=KMeans(n_clusters=5)
kmeans.fit(x)
print(davies_bouldin_score(x,kmeans.labels_))
kmeans=KMeans(n_clusters=4)
kmeans.fit(x)
print(davies_bouldin_score(x,kmeans.labels_))

#so the optimal no of clusters = 4
```
